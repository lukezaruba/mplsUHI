## 1. Data Processing and Exploration

In the following code blocks, the data will be downloaded, processed, and briefly explored. All necessary Python modules will also be imported.

```python
# Importing Modules
import pandas as pd
import numpy as np
import rasterio as rio
import geopandas as gpd
import folium as flm
import ipyleaflet as ipl

from rasterio.plot import show
import matplotlib.pyplot as plt
```

### LST Rasters

```python
# Read in LST Rasters
b10_20210618 = rio.open("/Users/lukezaruba/Desktop/SpatialDBGit/mplsUHI/LC08_L2SP_027029_20210618_20210628_02_T1_ST_TRAD.TIF")
b10_20210720 = rio.open("/Users/lukezaruba/Desktop/SpatialDBGit/mplsUHI/LC08_L2SP_027029_20210720_20210729_02_T1_ST_TRAD.TIF")
b10_20210922 = rio.open("/Users/lukezaruba/Desktop/SpatialDBGit/mplsUHI/LC08_L2SP_027029_20210922_20210930_02_T1_ST_TRAD.TIF")

# Printing Metadata
print("Image b10_20210618: " + "\n" + str(b10_20210618.meta) + "\n" + "\n" +
      "Image b10_20210720: " + "\n" + str(b10_20210720.meta) + "\n" + "\n" + 
      "Image b10_20210922: " + "\n" + str(b10_20210922.meta))
```

```python
# Plotting Raster Images
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(21,7))

show((b10_20210618, 1), ax=ax1, cmap='viridis', title="b10_20210618")
show((b10_20210720, 1), ax=ax2, cmap='viridis', title="b10_20210720")
show((b10_20210922, 1), ax=ax3, cmap='viridis', title="b10_20210922")

# Displaying Rasters
plt.show()
```

```python
# Find Average Values of Cells
b10_sum = np.add(b10_20210618.read(1), b10_20210720.read(1), b10_20210922.read(1))

b10_average = b10_sum / 3

# Plotting Histogram of Average LST Raster
plt.hist(b10_average, 5, density=1, facecolor='g', alpha=0.75)
plt.xlabel("LST")
plt.ylabel("Count")
plt.title('Histogram of LST')
plt.grid(True)
plt.show()
```
