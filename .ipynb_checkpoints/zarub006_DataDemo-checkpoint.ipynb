### Zonal Statistics (TEST)
The following section is simply testing the ability to calculate zonal statistics (statistical measures for raster cells within the area of given vector features) using rasterstats. In this test, rather than using the actualy vector landcover dataset (which will take a long time to load and use), we are using the boundary for Hennepin County to calculate the zonal statistics for LST within the entire county.

```python
# Calculating Zonal Statistics
stats = zs("/Users/lukezaruba/Desktop/SpatialDBGit/mplsUHI/hennepin.shp",
           "/Users/lukezaruba/Desktop/SpatialDBGit/mplsUHI/LC08_L2SP_027029_20210618_20210628_02_T1_ST_TRAD.TIF",
           stats=['min', 'max', 'median', 'majority'])

for stat in stats:
    [print(str(i) + ": " + str(stat[i])) for i in stat]
```

```python
# Start of Creating Tool to Convert DN to LST

def generateBT(DN, Ml, Al, K1, K2):
    """Given a DN value and other necessary values from metadata, will convert to Brightness Temperature (BT).
    In future, would like to expand to full conversion to LST."""
    
    # Calculating TOA Reflectance
    TOA = (Ml * DN) + Al
    
    # Calculating BT
    BT = (K2 / (np.log(K1 / TOA) + 1)) - 273.15
```
